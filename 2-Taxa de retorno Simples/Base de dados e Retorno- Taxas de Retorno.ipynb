Retorno de carteira de ações

```python
dataset=pd.read_csv('../1-Visualização de dados Financeiros/criptos.csv') #leitura do csv
dataset
```

```python
dataset_normalizado=dataset.copy()  #criei uma váriável que vai receber uma cópia do dataset
for i in dataset_normalizado.columns[1:]: #acessamos as colunas na posição 1 até a ultima coluna
    dataset_normalizado[i]=(dataset_normalizado[i]/dataset_normalizado[i][0])
```

```python
dataset_normalizado #apresentando gráfico normalizado começando com o preço na posição 1
```

```python
dataset_normalizado.plot(x='Date',figsize=(15,7)); #gerando gráfico normalizado
```

```python
dataset_normalizado.drop(labels=['Date'], axis=1, inplace=True) #apaga a coluna 1 que é a "Data" e inplace=True para fazer a alteração diretamente nessa base de dados.
```

```python
#temos as respectivas taxas de retorno para cada uma das ações e não temos mais a coluna 'Data'.
retorno_carteira=(dataset_normalizado/dataset_normalizado.shift(1))-1
retorno_carteira.head()  
```

```python
retorno_anual=retorno_carteira.mean()*365 #calculo do retorno médio anual dessa carteira
retorno_anual
```

```python
retorno_anual=retorno_anual*100 #calculo do retorno anual de cada ação
retorno_anual
```

```python
pesos_carteira2=np.array([28.59,20.82,19.43,9.51,8.38,6.64,6.63,0.0])
```

```python
pesos_carteira2.sum()
```

```python
#cálculo de retorno da carteira
#é feito a multiplicação do preço da ação em relação ao seu peso, e posteriormente é feito a soma.
np.dot(retorno_anual,pesos_carteira2)
```

Gráfico de Carteira X BOVA

```python
dataset=pd.read_csv('../1-Visualização de dados Financeiros/criptos.csv')
```

```python
dataset_normalizado = dataset.copy()
for i in dataset_normalizado.columns[1:]:
    dataset_normalizado [i]= (dataset_normalizado[i]/dataset_normalizado[i] [0])
dataset_normalizado.head()
```

```python
dataset_normalizado ['CARTEIRA']= (dataset_normalizado['MANA']+dataset_normalizado['POLKADOT']+dataset_normalizado['CARDANO']+dataset_normalizado['TRON']+dataset_normalizado['RIPLE']+dataset_normalizado['SAND']+dataset_normalizado['POLYGON']) / 7
dataset_normalizado
```

```python
figura=px.line(title='Comparativo Carteira x Bitcoin')
for i in dataset_normalizado.columns[1:]:
    figura.add_scatter(x=dataset_normalizado['Date'], y=dataset_normalizado[i], name=i)
figura.show()
```
