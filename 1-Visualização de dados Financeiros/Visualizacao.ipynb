```python
import pandas as pd
import numpy as np
from pandas_datareader import data as wb
import matplotlib.pyplot as plt 
import seaborn as sns
import plotly.express as px
import yfinance as yf
```

```python
cripto = ['MANA-USD', 'DOT-USD','ADA-USD','TRX-USD','XRP-USD','SAND-USD','MATIC-USD','UNI1-USD','BTC-USD']
```

```python
cripto_df= pd.DataFrame()
for cripto in cripto:
  cripto_df[cripto]= yf.download (cripto, data_source='yahoo', start='2020-01-01') ['Close']
```
